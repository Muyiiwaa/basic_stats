```python
df = orders_df[['unitprice', 'quantity','freight']].corr(method = 'pearson')
df2 = orders_df[['unitprice', 'quantity', 'freight']].corr(method = 'spearman')

fig, axes = plt.subplots(1,2, figsize = (20,10))
sns.heatmap(data = df, annot= True, fmt= '.2f', cmap= 'crest', ax=axes[0], cbar= False)
sns.heatmap(data = df2, annot= True, fmt= '.2f', ax=axes[1], cbar= False)

axes[0].set_title('Pearson Correlation')
axes[1].set_title('Spearman Rank Correlation')
```

```python
corr_value, p_value = stats.pearsonr(orders_df['unitprice'], orders_df['quantity'])


# DECISION RULE
null_hypothesis = 'There is no significant relationship'
alternative_hypothesis= 'There is a significant relationship'

alpha = 0.05


if p_value <= alpha:
    print(f'{alternative_hypothesis} @ p_value = {p_value}')
else:
    print(f'{null_hypothesis},  @ p_value = {p_value}')
```
